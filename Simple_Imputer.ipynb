```python
#Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer 
pd.set_option("display.max_columns", 100)
```

```python
#Load our data
fpath="/content/drive/MyDrive/CodingDojo/02-MachineLearning/Week05/Data/ames-housing-dojo-for-ml.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
#Setting PID as index
df = df.set_index("PID")
df.head()
```

### Define X and y



```python
# The target we are trying to predict
y = df['SalePrice']
# The features we will use to make the prediction
X = df.drop(columns = 'SalePrice')
```

### Train Test Split

```python
# Train test split 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.head()
```

## Imputation of Numeric Features with Median


#### 1) Create a list of features

```python
# Get the list of numeric column names 
num_cols = X_train.select_dtypes("number").columns
num_cols
```

#### 2) Instantiate the Imputer



```python
# Instantiate the imputer object from the SimpleImputer class with strategy 'median'
impute_median = SimpleImputer(strategy='median')
```

#### 3) Fit the Imputer of the Training Data



```python
# Fit the imputer object on the numeric training data with .fit() 
impute_median.fit(X_train[num_cols])
```

```python
# View the strategy used
impute_median.strategy
```

```python
# View the median values calculated during the .fit step
impute_median.statistics_
```

```python
# We can confirm that these are indeed the medians
X_train[num_cols].median().values
```

#### 4a) Transform the Training Data
Filling in the nulls for X_train set

```python
# Still have null values in numeric columns of **training data
null_count = X_train[num_cols].isna().sum()
null_count[ null_count>0]
```

```python
# Still have null values in numeric columns of **testing data
null_count = X_test[num_cols].isna().sum()
null_count[ null_count>0]
```

To actually alter these values with our imputer, we need to use .transform(). 

```python
# Still have null values in numeric columns of testing data
null_count = X_test[num_cols].isna().sum()
null_count[ null_count>0]
```

```python
# Transform the training data (fill in nulls with median)
X_train_num_imputed = impute_median.transform(X_train[num_cols])
X_train_num_imputed
```

#### 4b) Transform the Test Data

```python
# Check for null values in numeric columns of test set
X_test[num_cols].isna().sum().sum()
```

```python
# Transform the testing data (fill in nulls with median from training data)
X_test_num_imputed = impute_median.transform(X_test[num_cols])
X_train_num_imputed.head()
```

#### Putting it All Together



```python
# Define the list of features for this strategy
num_cols = X_train.select_dtypes("number").columns
# Instantiate the imputer object from the SimpleImputer class with strategy 'median'
impute_median = SimpleImputer(strategy='median')
# Fit the imputer object on the numeric training data with .fit() 
impute_median.fit(X_train[num_cols])
# Transform the training data
X_train_num_imputed = impute_median.transform(X_train[num_cols])
# Transfrom the testing data
X_test_num_imputed = impute_median.transform(X_test[num_cols])
```

## Imputation of Categorical and Ordinal Features with Placeholder

#### 1) Define the List of Features



```python
# Get the list of categorical and ordinal column names 
cat_cols = df.select_dtypes('object').columns
cat_cols
```

#### 2) Instantiate the Imputer



```python
# Instantiate the new imputer
impute_na = SimpleImputer(strategy="constant", fill_value="NA")
```

#### 3) Fit the Imputer on the Training Data

```python
# Fit the imputer on the training data
impute_na.fit(X_train[cat_cols])
```

#### 4a) Transform the Training Data



```python
# Check for missing values in cat columns
X_train[cat_cols].isna().sum().sum()
```

```python
# Transform the categorical training data
X_train_cat_imputed = impute_na.transform(X_train[cat_cols])
```

```python
# Confirm that all null vales in cat columns are filled in training set
X_train_cat_imputed.isna().sum().sum()
```

#### 4b) Transform the Testing Data



```python
# Transform the categorical testing data
X_test_cat_imputed = impute_na.transform(X_test[cat_cols])
```

```python
# Confirm all missing values in test set are filled
X_test_cat_imputed.isna().sum().sum()
```

#### Putting it All Together

```python
# Define list of features for this strategy
cat_cols = df.select_dtypes("object").columns
# Instantiate the imputer with the desired strategy
impute_na = SimpleImputer(strategy="constant", fill_value="NA")
# Fit the imputer object on the training data with .fit
impute_na.fit(X_train[cat_cols])
# Transform the categorical training data
X_train_cat_imputed = impute_na.transform(X_train[cat_cols])
X_test_cat_imputed = impute_na.transform(X_test[cat_cols])
```

```python
#Checking to see if X_train categories nulls have been filled
X_train_cat_imputed.isna().sum().sum()
```

```python
#Checking to see if X_train categories nulls have been filled
X_test_cat_imputed.isna().sum().sum()
```
