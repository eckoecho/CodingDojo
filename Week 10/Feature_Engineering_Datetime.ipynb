```python
import pandas as pd
df2 = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSrgrUnz8mdosU-_k0aECouymqwds_mlaHpYlXzRtf7MBJ4N1r1inCfSDebaXwTVfLtH133EhwKf3mi/pub?gid=394699239&single=true&output=csv',                  usecols=['date','price','bedrooms','bathrooms'])
df2.head()
```

```python
#Checking data type for our date column
df2.info()
```

In this case, the 'date' column loaded as an 'object' type. We can use pd.to_datetime() to change the time to datatime.

```python
#Using .to_datetime to convert columns datatype to datetime.
df2["date"] = pd.to_datetime(df2["date"])
df2.info()
```

```python
#Pulling year from our date and create a new column
df2["year"] = df2["date"].dt.year
#Pulling month from date and making a new column
df2["month (numeric)"] = df2["date"].dt.month
#Getting month name from date and making new column
df2["month (name)"] = df2["date"].dt.month_name()
#Getting day of the month from date and making new column
df2["day of month"] = df2["date"].dt.day
#Getting numeric day of the week and creating new column
df2["day of the week (numeric)"] = df2["date"].dt.weekday
#Getting name of the weekday and creating a new column
df2["day of the week (name)"] = df2["date"].dt.day_name()
df2.head()
```
