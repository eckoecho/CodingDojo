```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/CodingDojo/03-AdvancedML/Week10/Data/bikeshare_train - bikeshare_train.csv")
df.head()
```

import the data the drop the 'casual' and 'registered' columns. These are redundant with your target, 'count'.

```python
#Dropping 2 redundant columns
df.drop(columns=["casual", "registered"], inplace=True)
df.head()
```

```python
df.info()
```

In this case, the 'date' column loaded as an 'object' type. We can use pd.to_datetime() to change the time to datatime.

```python
df["datetime"] = pd.to_datetime(df["datetime"])
```

Transform the 'datetime' column into a datetime type and use it to create 3 new columns in the data frame.

```python
#Getting month from datetime and creating new column
df["month (name)"] = df["datetime"].dt.month_name
#Getting date of the week from datetime and creating new column
df["day of the week (name)"] = df["datetime"].dt.day_name()
#Getting the hour of the day from datetime column and creating new column
df["hour of the day"] = df["datetime"].dt.hour
```

```python
#Checking datatypes after creating new columns
df.info()
```

```python
#Converting "hour of day" to object
df["hour of the day"] = df["hour of the day"].astype(object)
df.info()
```

```python
#Dropping columns i no longer need
df.drop(columns=["datetime", "season"], inplace=True)
df.head()
```

The temperatures in the 'temp' and 'atemp' columns are in Celsius. Use `.apply()` and a Lambda function to convert them to Fahrenheit.

Forumla is: (°C * 1.8) + 32 = °F

```python
#Creating new columns when applying lambda function
df["temp"] = df["temp"].apply(lambda c: float (c*1.8)+32)
df["atemp"] = df["atemp"].apply(lambda c: float (c*1.8)+32)
df.head()
```

which shows how much warmer or colder the current temperature ('temp') is than the average temperate for that day of the year ('atemp')

```python
df["temp_variance"] = df["atemp"] - df["temp"]
df["temp_variance"].describe()
```

```python
#Dropping column i no longer need
df.drop(columns="atemp", inplace=True)
df.head()
```
