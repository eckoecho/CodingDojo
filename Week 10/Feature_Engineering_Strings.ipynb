```python
import pandas as pd
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vReZBM5OC6GLYbacisp_ToNiu3CLWxqPXw7mWBsdRjnYOFLWNufdQ4qd8u5qTzUF2_sBUAMEi5cgy1U/pub?gid=1040198428&single=true&output=csv')
df.head()
```

# Separating Features: Strings

```python
# create 2 new columns, FirstName and LastName by splitting the Name column
df[["LastName", "FirstName"]] = df["Name"].str.split(",", expand=True)
#Dropping the Name column since we create a new column from it
df.drop(columns="Name")
df.head()
```

# Clean-up

```python
#Looking at whitespace accidently left when splitting
df.loc[0,'FirstName']
```

```python
#Erases whitespace from front and back of a string
df["FirstName"] = df["FirstName"].str.strip()
```

# Combining Strings

```python
#Adding columns back together but with a space between first and last name
df["Name"] = df["FirstName"]+ " " + df["LastName"]
#Dropping columns we just combined
df.drop(columns=["FirstName", "LastName"], inplace=True)
df.head()
```
