```python
#import data
import pandas as pd
```

```python
#Load data
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vReZBM5OC6GLYbacisp_ToNiu3CLWxqPXw7mWBsdRjnYOFLWNufdQ4qd8u5qTzUF2_sBUAMEi5cgy1U/pub?gid=1040198428&single=true&output=csv')
df.head()
```

# Summing Features

```python
#Summing the values from 2 columns and creating a new column
df["TotalFamily"] = df["SibSp"] + df["Parch"]
#dropping old columns since we converged columns
df = df.drop(columns=["SibSp", "Parch"])
df.head()
```

# Concatenating Features

```python
#Rounding Ages to nearest decade
df["Age"] = df["Age"].round(-1)
df.head()
```

```python
#Creating a new column by combining the Sex and Age column
#Age column must be turned into a string to combing with another string column
df["GenderAge"] = df["Sex"] + df["Age"].astype(str)
#Droping sex and age column since we combined them
df.drop(columns=["Age", "Sex"], inplace=True)
df.head()
```

# Squaring and Multiplying Features

```python
#Calculating a new column by multiplying and squaring PClass
df["NormedFare"] = df["Fare"] * df["Pclass"] ** 2
#Dropping the columns we already combined
df.drop(columns=["Fare", "Pclass"], inplace=True)
df.head()
```
