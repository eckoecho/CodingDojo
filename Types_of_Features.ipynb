```python
# Import packages and load data
import pandas as pd
from sklearn.model_selection import train_test_split
pd.set_option("display.max_columns", 100)
```

```python
## Load the data and set the index
fpath = "/content/drive/MyDrive/CodingDojo/02-MachineLearning/Week05/Data/Prepped-ames-housing-dojo-for-ml.csv"
df = pd.read_csv(fpath)
df = df.set_index("PID")
df.head()
```

## Format for ML

```python
#Setting X to be the feature set
X = df.drop(columns=["SalePrice"]).copy()
#Setting y to be the target
y = df["SalePrice"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Replace the Ordinal feature (Exter Qual) with Numbers

```python
# display categories in ExterQual
X_train['Exter Qual'].value_counts(dropna=False)
```

```python
# define dictionary to replace ordinal categories
quals = {"Po":0, "Fa":1, "TA":2, "Gd":3, "Ex":4}
# apply the dictionary to the column in the train set
X_train["Exter Qual"] = X_train["Exter Qual"].replace(quals)
X_test["Exter Qual"] = X_test["Exter Qual"].replace(quals)
# view the dataframe to make sure it worked
X_train["Exter Qual"].value_counts(dropna=False)
```
